```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
corr_same = np.load('positional_corr_same.npy')
csame_avg = np.hstack((1,np.mean(corr_same,axis=1)))
corr_different = np.load('positional_corr_diff.npy')
cdiff_avg = np.mean(corr_different, axis=1)
```

```python
plt.plot([0,1,2,3], csame_avg, marker='o', label='Same item')
plt.plot([0,1,2,3], cdiff_avg, marker='o', label='Non-confusable')
plt.xticks([0,1,2,3])
plt.ylabel("Correlation", fontsize=16)
plt.xlabel("Distance", fontsize=16)
plt.legend()
```
